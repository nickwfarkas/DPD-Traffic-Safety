```python
import json
import re
import csv
import os
```

```python
# Load the HTML content
for i in range(88,145):
    file_name = f'./data/view-source_https___www.concertarchives.org_locations_detroit-mi_date=past&page={i}.html'
    with open(file_name, "r", encoding="utf-8") as file:
        html = file.read()

    # Extract JSON content from <script type="application/ld+json">
    json_blocks = re.findall(r'<script type="application/ld\+json">\s*(\{.*?\})\s*</script>', html, re.DOTALL)

    concerts = []

    for block in json_blocks:
        try:
            data = json.loads(block)
            if "event" in data:
                for event in data["event"]:
                    concert_info = {
                        "Date": event.get("startDate", ""),
                        "Concert": event.get("name", ""),
                        "Venue": event.get("location", {}).get("name", ""),
                        "Location": event.get("location", {}).get("address", "")
                    }
                    concerts.append(concert_info)
        except json.JSONDecodeError as e:
            print("Error decoding JSON block:", e)

    # Save or append to CSV
    csv_file = "concerts.csv"
    file_exists = os.path.isfile(csv_file)

    with open(csv_file, "a", newline='', encoding="utf-8") as csvfile:
        fieldnames = ["Date", "Concert", "Venue", "Location"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        # Write header only if file doesn't already exist
        if not file_exists:
            writer.writeheader()

        writer.writerows(concerts)

    print(f"Appended {len(concerts)} concerts to {csv_file}")
```
